```python
import pandas as pd
```

```python
train_df = pd.read_csv('challenge_train.csv')
print('Lenght: ',len(train_df))
train_df.head()
```

```python
dummies_type_df = pd.get_dummies(train_df['type'])
dummies_god_df = pd.get_dummies(train_df['god'])
dummies_god_df
```

```python
processed_train_df = train_df.join(dummies_type_df)
processed_train_df = processed_train_df.join(dummies_god_df)
processed_train_df = processed_train_df.drop(['id','name','type','god'], axis = 1)
#base pronta pra treinamento
processed_train_df.head()
```

```python
processed_train_df['strategy'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sn
from sklearn.metrics import classification_report

y = processed_train_df['strategy']
X = processed_train_df.drop('strategy', axis = 1)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)

logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,y_train)
y_pred=logistic_regression.predict(X_test)

confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])

print(confusion_matrix)

print(classification_report(y_test, y_pred))
```
